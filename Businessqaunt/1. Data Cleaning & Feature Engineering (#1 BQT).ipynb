```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
from shortcuts import data_loader
import numpy as np
```

```python
filename = "1. AssignmentBusinessQuant.csv"
data = pd.read_csv(data_loader.data_path(filename))
data["Sales Last Year"] = data["Sales Growth %"].values
```

```python
data.head()
```

```python
tdata = data.groupby(["Item", "Year"]).sum()
tdata
```

```python
for item in tdata.index.get_level_values(0).unique():
#     print(item)
    tdata.loc[item, 'Sales Last Year'] = tdata.loc[item, 'Sales'].shift().values
    tdata.loc[item, 'Sales Growth %'] = np.round(
        (tdata.loc[item, "Sales"] / tdata.loc[item, "Sales Last Year"] - 1) * 100,
        1).values
```

```python
tdata["Year"] = tdata.index.get_level_values(1)
tdata.index = tdata.index.get_level_values(0)
tdata.reset_index(inplace=True)
tdata = tdata[["Item", "Year", "Sales", "Sales Last Year", "Sales Growth %"]]
```

```python
tdata.to_csv(data_loader.data_path(filename, "Results"), index=False)
```
